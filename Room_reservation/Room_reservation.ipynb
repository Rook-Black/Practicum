# Прогнозирование оттока клиентов в сети отелей «Как в гостях»

Заказчик этого исследования — сеть отелей «Как в гостях».
Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.  

Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

# Бизнес-метрика и другие данные

Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.
В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.  

Стоимость номеров отеля: 
- категория A: за ночь — 1 000, разовое обслуживание — 400;
- категория B: за ночь — 800, разовое обслуживание — 350;
- категория C: за ночь — 600, разовое обслуживание — 350;
- категория D: за ночь — 550, разовое обслуживание — 150;
- категория E: за ночь — 500, разовое обслуживание — 150;
- категория F: за ночь — 450, разовое обслуживание — 150;
- категория G: за ночь — 350, разовое обслуживание — 150.  

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.
Убытки отеля в случае отмены брони номера — это стоимость одной уборки и одной ночи с учётом сезонного коэффициента.
На разработку системы прогнозирования заложен бюджет — 400 000. При этом необходимо учесть, что внедрение модели должно окупиться за тестовый период. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

- total_of_special_requests — количество специальных отметок.

Теперь создадим функцию для просчета количества дней прожитых человеком. Это нужно что бы учесть перехода одного сезона к другому.

```python
def season (df):
    
    leap_year = 0
    
    #Проверяем на високосность год и является ли месяц февралем.
    if df['arrival_date_year'] % 4 == 0 and df['arrival_date_month'] == "February": leap_year = 1
    
    #Проверяем будет ли челоек встречать новый месяц в отеле
    if (df["total_nights"] + df["arrival_date_day_of_month"] + leap_year >
        (seasonal_factor
         .loc[seasonal_factor['month']==
              df['arrival_date_month'],
              'days']
         .values[0])):
        
        #Получаем дни прожитые в текущем месяце умноженые на коэффициент сезона
        day = (((seasonal_factor
                 .loc[seasonal_factor['month']==
                      df['arrival_date_month'],
                      'days']
                 .values[0])+
               leap_year-
               df["arrival_date_day_of_month"]+
               1)*
               (seasonal_factor
                .loc[seasonal_factor['month']==
                     df['arrival_date_month'],
                     'coefficient'])
               .values[0])
        
        #Переносим оставшиеся дни на следующий месяц
        df['total_nights'] = (df["total_nights"]+
                              df["arrival_date_day_of_month"]-
                              (seasonal_factor
                               .loc[seasonal_factor['month']==
                                    df['arrival_date_month'],
                                    'days']
                               .values[0])-
                              leap_year-
                             1)
        #Изменяем дату заезда на перове число
        df['arrival_date_day_of_month'] = 1
        
        #Проверяем закончился ли год
        if df['arrival_date_month'] == "December": 
            df['arrival_date_month'] = "January"
            df['arrival_date_year'] = df['arrival_date_year'] + 1
            
        #Есдли не закончился просто меняем месяц на слудующий
        else: df['arrival_date_month'] = (seasonal_factor
                                          .loc[seasonal_factor
                                               .loc[seasonal_factor['month']==
                                                    df['arrival_date_month']]
                                               .index+1,
                                               "month"]
                                          .values[0])
        
        #Возвращаем количество дней прожитых в текущем месяце и снова вызываем функцию для подсчета прожитых дней в следующем месяце
        return day + season(df)
    else:
        
        #Новый месяц человек не встретит, следовательно можно просто вернуть количество дней умноженых на коэффициент сезона 
        day = (df['total_nights']*
               (seasonal_factor
                .loc[seasonal_factor['month']==
                     df['arrival_date_month'],
                     'coefficient'])
               .values[0])
        return day
```

Теперь создаем функцию для посчета прибыли.

```python
def profit (df):
    #Вводим переменную для подсчета прибыли
    pay = 0
    
    #Переменная для подсчета количества уборок в номере
    service_amount = 1
    
    #Проверяем была ли отмена
    if df["is_canceled"] == 0:
        
        #Если не было, то производим подсчет прибыли
        #Считаем количество уборок номера
        #if (df["total_nights"]) >= 2: 
        service_amount += (df["total_nights"]) // 2
        
        #Подсчитываем доход с номера. Стоимость номера умножается на количество дней прожитых в каждом сезоне и вычитается уборка номера.
        pay += ((rooms.loc[rooms['type'] == df['reserved_room_type'], "cost"])*
                season(df)-
                service_amount*
                (rooms.loc[rooms['type'] == df['reserved_room_type'], "service"]))
    else:
        #Отмена идет в минус, подсчет стоимость номера плюс уборка.
        pay -= (rooms.loc[rooms['type'] == df['reserved_room_type'], "cost"]*
                seasonal_factor.loc[seasonal_factor['month'] == df['arrival_date_month'], 'coefficient'].values[0]+
                rooms.loc[rooms['type'] == df['reserved_room_type'], "service"])
        
    return pay
```

##### Корректировка функции подсчета.

По скольку нужно теперь учитывать предсказание модели, нужно подкорректировать функцию.

```python
def profit_ml (df, model = ''):
    #Вводим переменную для подсчета прибыли
    pay = 0
    
    #Добавим переменную бронирования
    booking = 0
    if model != '':
        if df[f'canceled_{model}'] == 1:
            booking += ((rooms.loc[rooms['type'] == df['reserved_room_type'], "cost"]+
                         rooms.loc[rooms['type'] == df['reserved_room_type'], "service"])*
                        .8*
                        seasonal_factor.loc[seasonal_factor['month'] == df['arrival_date_month'], 'coefficient'])
    
    #Переменная для подсчета количества уборок в номере
    service_amount = 1
    
    #Проверяем была ли отмена
    if df["is_canceled"] == 0:
        
        #Если не было, то производим подсчет прибыли
        #Считаем количество уборок номера
        #if (df["total_nights"]) >= 2: 
        service_amount += (df["total_nights"]) // 2
        
        #Подсчитываем доход с номера. Стоимость номера умножается на количество дней прожитых в каждом сезоне и вычитается уборка номера.
        pay += ((rooms.loc[rooms['type'] == df['reserved_room_type'], "cost"])*
                season(df)-
                service_amount*
                (rooms.loc[rooms['type'] == df['reserved_room_type'], "service"]))
    else:
        #Отмена идет в минус, подсчет стоимость номера плюс уборка.
        pay -= (rooms.loc[rooms['type'] == df['reserved_room_type'], "cost"]*
                seasonal_factor.loc[seasonal_factor['month'] == df['arrival_date_month'], 'coefficient'].values[0]+
                rooms.loc[rooms['type'] == df['reserved_room_type'], "service"]-
                booking)
    return pay
```

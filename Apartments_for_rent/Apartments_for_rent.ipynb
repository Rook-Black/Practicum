## Предобработка данных

Первым делом проверим в каких столбцах встречаются пропуски и попробуем оценить их прчину.

```python
df.isna().sum()
```

ceiling_height, floors_total, living_area, kitchen_area, locality_name в данных столбцах есть пропуски скорее всего из-за не знания людей данной информации, заменить их чем либо вряд ли получится.

airport_nearest, cityCenters_nearest возможно квартиры находтся в поселках или деревнях, где нет ни центра, ни аэропрота.

is_apartmens, balcony, parks_around3000, pounds_around3000 в даных столбцах отсутствуют зачения поскольку отсутствуют данные объекты в/рядом с квартирой, следовательно люди пропускали данный пункт. Можно заполнить данные пропуски нулями.

parks_nearest, pounds_nearest тут совокупность двух вышестоящих факторов, следовательно часть данных можно восстановить, используя пропуски и значения в столбцах parks_around3000, pounds_around3000 

Займемся пропусками и при необходимости меняем типы столбцов.
По скольку столбец "is_apartment" принимает только два значения логичнее использовать булевые значения

```python
df["is_apartment"] = df["is_apartment"].fillna(False).astype(bool)
```

Проверим гипотезу про аэропорты и центры.

```python
df.loc[df["airports_nearest"].isna(), ["locality_name"]].value_counts().head(20)
```

```python
df.loc[df["cityCenters_nearest"].isna(), ["locality_name"]].value_counts().head(20)
```

По большей части предположение было правильное, поэтому заменим все пропуски нулевыми значениями и переведем все в целые числа. Заодно в целые занчения переведем количество балконов, парков и прудов, поскольку не может быть половины чего-либо из выше приведенного.

```python
for column in ["balcony", "parks_around3000", "ponds_around3000", "cityCenters_nearest", "airports_nearest"]:
    df[column] = df[column].fillna(0).astype(int)
```

Так же растояние в метрах слишком точное, в км будет достаточно, по сколкьку чаще всего добираются до этих мест на транспорте, а вот до парков и озер лучше оставить в метрах

```python
df["cityCenters_nearest"] = (df["cityCenters_nearest"] / 1000).round()
df["airports_nearest"] = (df["airports_nearest"] / 1000).round()
```

Теперь можно поработать со столбцом "parks_nearest", возьмем столец "parks_around3000" и там где нет парков поставим 0

```python
df.loc[df["parks_around3000"] == 0,["parks_nearest"]] = 0
df["parks_nearest"].isna().sum()
```

Как видим все пропуски были связаны с отсутствием парков в районе и люди просто пропускали данные.

Проделаем все то же самое и для прудов.

```python
df.loc[df["ponds_around3000"] == 0,["ponds_nearest"]] = 0
df["ponds_nearest"].isna().sum()
```

Так же для столбцa floors_total будем использовать целые числа, следовательно переменные с плавющей точкой в них не нужны. Предварительно заполнив пропуски, будем считать что квартира изначально находится на последнем этаже и запишем это в столбец. 

```python
df["floors_total"] = df["floors_total"].fillna(df["floor"])
df["floors_total"] = df["floors_total"].astype(int)
```

```python
df.isna().sum()
```

```python
df.info()
```

### Вывод

В основном все пропуски были сделаны по человеческому фактору, при отсутствии того или иного фактора или просто пропускание строк анкеты при заполнении. так же встертились пропуски в столбцах заполняемых автоматически, тут просто не предусмотрели момент отсутствия пурод или аэропортов в городе из-за чего программа не могла вставить необходимое значение.
В результате имеем почти идеально подготовленную таблицу для работы, все столбцы приведены к необходимым переменным и по возможности заполнены пропуски

## Расчёты и добавление результатов в таблицу

Для упрощения работы с таблице дабавим некоторые столбцы:
1. цена за квадратный метр "price_by_metr"
2. день недели, месяц и год публикации объявления "weekday", "year", "month"
3. этаж квартиры; варианты — первый, последний, другой "floor_name"
4. соотношение жилой и общей площади, а также отношение площади кухни к общей "ratio", "ratio_kitchen"

```python
df["price_by_metr"] = df["last_price"] / df["total_area"]
```

Для работы с датой переведем сначала столбец "first_day_exposition" в необходимый вид.

Так же для удобной работы создадим словарь с этажами.

```python
df["first_day_exposition"] = pd.to_datetime(df["first_day_exposition"], format = '%Y-%m-%dT%H:%M:%S')
df['weekday'] = df["first_day_exposition"].dt.weekday
df["year"] = df["first_day_exposition"].dt.year
df["month"] = df["first_day_exposition"].dt.month

floors = {
    0: "первый",
    1: "другой",
    2: "последний"
}

df.loc[df["floor"] == 1, ["floor_name"]] = 0
df.loc[df["floor"] == df["floors_total"], ["floor_name"]] = 2
df["floor_name"] = df["floor_name"].fillna(1)
df["floor_name"] = df["floor_name"].astype(int)

df["ratio"] = df["living_area"]/df["total_area"]
df["ratio_kitchen"] = df["kitchen_area"]/df["total_area"]

df.sample(5)
```
